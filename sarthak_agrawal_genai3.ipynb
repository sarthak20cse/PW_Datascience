# Question 1: What are Large Language Models (LLMs) and how do they function?
# Answer:

Large Language Models (LLMs) are advanced artificial intelligence systems trained on massive amounts of text data to understand, generate, summarize, translate, and answer questions in natural language. They are based on deep learning architectures, primarily the Transformer architecture, which enables them to process and understand context within large text sequences.

How LLMs Function
Data Collection
LLMs are trained using large datasets containing books, articles, websites, research papers, and other textual sources.

Tokenization
Text is broken into smaller units called tokens (words, subwords, or characters).

Training Process
The model learns patterns by predicting the next token in a sequence.
Billions of parameters are adjusted through optimization techniques.

Transformer Architecture
Uses self-attention mechanisms to understand relationships between words.
Captures long-range dependencies in text.

Inference
When a user provides a prompt, the model predicts the most probable sequence of words to generate a response.

Question 7: Integrate LangChain with a third-party API (e.g., weather, news) and show how responses can be generated via LLMs.

````python
import requests
from langchain_groq import ChatGroq

WEATHER_API_KEY = "9e1e7****"
city = "Mumbai"

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    groq_api_key=GROQ_KEY
)

try:
    url = f"http://api.weatherapi.com/v1/current.json?key={WEATHER_API_KEY}&q={city}"

    data = requests.get(url).json()

    if "current" in data:
        temperature = data["current"]["temp_c"]
        description = data["current"]["condition"]["text"]
    else:
        print("API Error:", data)
        temperature = "30"
        description = "Partly Cloudy"

    prompt_text = f"""
    Weather in {city}:
    Temperature: {temperature}°C
    Condition: {description}

    Summarize this weather for a traveler.
    """

    response = llm.invoke(prompt_text)

    print(response.content)

except Exception as e:
    print("Error:", e)
````

Question 8: Create a LlamaIndex implementation that indexes a local text file and retrieves answers from it.

````python
# ==========================================================
# STEP 1: Install Required Libraries
# ==========================================================

!pip install -q llama-index
!pip install -q llama-index-llms-groq
!pip install -q llama-index-embeddings-huggingface


# ==========================================================
# STEP 2: Create sample.txt
# ==========================================================

sample_text = """
LlamaIndex is a framework that connects Large Language Models (LLMs)
with external data sources.

It provides tools for document loading, indexing, retrieval, and querying.

Groq is a high-performance AI inference platform that enables fast execution
of large language models such as Llama 3.

Retrieval-Augmented Generation (RAG) is a technique that combines
information retrieval with language generation.

The RAG workflow consists of:

1. Loading documents
2. Creating embeddings
3. Building an index
4. Retrieving relevant information
5. Generating answers using an LLM

Benefits of LlamaIndex:

- Easy document ingestion
- Efficient retrieval
- Supports multiple vector databases
- Simplifies AI application development

LlamaIndex and Groq can be used together to create fast and scalable
question-answering systems.
"""

with open("sample.txt", "w") as f:
    f.write(sample_text)

print("sample.txt created successfully!")


# ==========================================================
# STEP 3: Import Libraries
# ==========================================================

import os
from getpass import getpass

from llama_index.core import (
    VectorStoreIndex,
    SimpleDirectoryReader,
    Settings
)

from llama_index.llms.groq import Groq
from llama_index.embeddings.huggingface import HuggingFaceEmbedding


# ==========================================================
# STEP 4: Enter Groq API Key
# ==========================================================

groq_api_key = getpass("Enter your Groq API Key: ")

os.environ["GROQ_API_KEY"] = groq_api_key


# ==========================================================
# STEP 5: Configure Groq LLM
# ==========================================================

Settings.llm = Groq(
    model="llama-3.3-70b-versatile",
    api_key=groq_api_key
)


# ==========================================================
# STEP 6: Configure Embedding Model
# ==========================================================

Settings.embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-small-en-v1.5"
)


# ==========================================================
# STEP 7: Load Document
# ==========================================================

documents = SimpleDirectoryReader(
    input_files=["sample.txt"]
).load_data()

print(f"Documents Loaded: {len(documents)}")


# ==========================================================
# STEP 8: Create Vector Index
# ==========================================================

index = VectorStoreIndex.from_documents(documents)

print("Index created successfully!")


# ==========================================================
# STEP 9: Create Query Engine
# ==========================================================

query_engine = index.as_query_engine()


# ==========================================================
# STEP 10: Ask Questions
# ==========================================================

questions = [
    "What is LlamaIndex?",
    "What is Groq?",
    "Explain the RAG workflow.",
    "What are the benefits of LlamaIndex?"
]

for question in questions:

    response = query_engine.query(question)

    print("\n" + "="*60)
    print("Question:")
    print(question)

    print("\nAnswer:")
    print(response)
````

Question 9: Demonstrate combining LangChain with LlamaIndex to create a simple document-based Q&A chatbot.

````python
# ==========================
# INSTALL DEPENDENCIES
# ==========================

!pip install -q llama-index==0.14.16
!pip install -q llama-index-llms-groq==0.5.0
!pip install -q llama-index-embeddings-huggingface==0.6.1
!pip install -q langchain==0.3.27
!pip install -q langchain-core==0.3.75

# ==========================
# CREATE SAMPLE DOCUMENT
# ==========================

sample_text = """
LangChain is an open-source framework used to build applications powered by Large Language Models.

LlamaIndex is a framework that connects external data sources with Large Language Models.
It provides document loading, indexing, retrieval, and querying capabilities.

Groq is a high-performance AI inference platform that enables ultra-fast execution of Large Language Models such as Llama 3.

A document-based chatbot retrieves information from documents and answers user questions based on document content.

Retrieval-Augmented Generation (RAG) combines information retrieval with language generation.

Benefits of RAG:
1. Reduces hallucinations.
2. Improves answer accuracy.
3. Uses external knowledge.
4. Provides context-aware responses.

LangChain manages chatbot workflows.
LlamaIndex manages document retrieval.
Groq generates final answers.
"""

with open("sample.txt", "w") as f:
    f.write(sample_text)

print("sample.txt created successfully.")

# ==========================
# IMPORT LIBRARIES
# ==========================

import os
from getpass import getpass

from llama_index.core import (
    VectorStoreIndex,
    SimpleDirectoryReader,
    Settings
)

from llama_index.llms.groq import Groq
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

from langchain_core.runnables import RunnableLambda

# ==========================
# ENTER GROQ API KEY
# ==========================

groq_api_key = getpass("Enter your Groq API Key: ")

os.environ["GROQ_API_KEY"] = groq_api_key

# ==========================
# CONFIGURE GROQ MODEL
# ==========================

Settings.llm = Groq(
    model="llama-3.3-70b-versatile",
    api_key=groq_api_key
)

# ==========================
# CONFIGURE EMBEDDINGS
# ==========================

Settings.embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-small-en-v1.5"
)

print("Models loaded successfully.")

# ==========================
# LOAD DOCUMENT
# ==========================

documents = SimpleDirectoryReader(
    input_files=["sample.txt"]
).load_data()

print(f"Loaded {len(documents)} document(s).")

# ==========================
# CREATE VECTOR INDEX
# ==========================

index = VectorStoreIndex.from_documents(documents)

query_engine = index.as_query_engine()

print("Vector index created successfully.")

# ==========================
# LANGCHAIN + LLAMAINDEX
# ==========================

def ask_document(question):
    response = query_engine.query(question)
    return str(response)

chatbot = RunnableLambda(ask_document)

print("LangChain and LlamaIndex integrated successfully.")

# ==========================
# SAMPLE QUESTIONS
# ==========================

sample_questions = [
    "What is LangChain?",
    "What is LlamaIndex?",
    "What is RAG?"
]

print("\nSample Outputs")
print("=" * 60)

for q in sample_questions:
    print(f"\nQuestion: {q}")
    print("Answer:", chatbot.invoke(q))

# ==========================
# INTERACTIVE CHATBOT
# ==========================

print("\n" + "=" * 60)
print("DOCUMENT-BASED Q&A CHATBOT")
print("=" * 60)

question_count = 1

while True:

    print(f"\nQuestion {question_count}")

    question = input("You: ")

    answer = chatbot.invoke(question)

    print("\nChatbot:", answer)

    choice = input(
        "\nDo you want to ask another question? (yes/no): "
    ).strip().lower()

    if choice != "yes":
        print("\nThank you for using the chatbot.")
        break

    question_count += 1
````
