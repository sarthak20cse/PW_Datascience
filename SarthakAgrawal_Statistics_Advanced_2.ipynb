 Question 1: What is Hypothesis Testing in Statistics?



**Hypothesis testing** is a statistical method used to make decisions or draw conclusions about a population based on sample data.

---

## üîπ Key Concepts

### **1. Null Hypothesis (H‚ÇÄ)**
The default assumption:  
**Nothing special is happening.**  
Example: *The new medicine has no effect.*

### **2. Alternative Hypothesis (H‚ÇÅ)**
The claim we want to test:  
**Something is happening.**  
Example: *The new medicine works.*

---

## üîπ Steps in Hypothesis Testing

1. **State H‚ÇÄ and H‚ÇÅ**
2. **Choose significance level (Œ±)** ‚Äî usually 0.05  
3. **Collect sample data**
4. **Perform a statistical test** (t-test, z-test, chi-square, etc.)
5. **Compute p-value**
6. **Decision rule:**
   - If **p ‚â§ Œ±**, reject H‚ÇÄ (evidence supports H‚ÇÅ)  
   - If **p > Œ±**, fail to reject H‚ÇÄ (insufficient evidence)

---

## ‚úîÔ∏è In one line:
Hypothesis testing determines whether sample data provides enough evidence to support a claim about a population.


Question 2: What is the Null Hypothesis, and how does it differ from the Alternative Hypothesis?



## üîπ Null Hypothesis (H‚ÇÄ)
The **null hypothesis** is the default assumption in statistics.  
It states that **there is no effect, no difference, or no relationship** in the population.

Example:  
*The new drug has no effect on blood pressure.*

---

## üîπ Alternative Hypothesis (H‚ÇÅ or Ha)
The **alternative hypothesis** is what you want to test or prove.  
It states that **there is an effect, a difference, or a relationship**.

Example:  
*The new drug lowers blood pressure.*

---

## üîπ Key Differences

| Feature | Null Hypothesis (H‚ÇÄ) | Alternative Hypothesis (H‚ÇÅ) |
|--------|------------------------|------------------------------|
| Meaning | No effect / no difference | Effect / difference exists |
| Role | Default assumption | Claim being tested |
| Objective | Try to reject it | Try to find evidence for it |
| Decision | Reject or fail to reject | Accept only if H‚ÇÄ is rejected |

---

## ‚úîÔ∏è In one line:
**H‚ÇÄ says "nothing changes."  
H‚ÇÅ says "something changes."**


 Question 3: Explain the Significance Level in Hypothesis Testing and Its Role in Deciding the Outcome of a Test



## üîπ What is the Significance Level (Œ±)?

The **significance level**, denoted by **Œ±**, is the threshold used in hypothesis testing to decide whether to reject the null hypothesis (H‚ÇÄ).  
It represents the **probability of making a Type I error**, which means:

**Rejecting H‚ÇÄ when it is actually true.**

Common values:  
- Œ± = 0.05 (most used)  
- Œ± = 0.01  
- Œ± = 0.10  

---

## üîπ Why is the Significance Level Important?

The significance level sets the **strictness** of the test.  
A lower Œ± means you need stronger evidence to reject H‚ÇÄ.

Example:  
- Œ± = 0.05 ‚Üí 5% chance of incorrectly rejecting H‚ÇÄ  
- Œ± = 0.01 ‚Üí 1% chance (more strict)  

---

## üîπ Role in Deciding the Outcome

After calculating the **p-value** from your sample data:

### Decision Rule:
- If **p ‚â§ Œ±** ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ)
- If **p > Œ±** ‚Üí Fail to reject H‚ÇÄ (insufficient evidence)

---

## ‚úîÔ∏è In one line:
The significance level (Œ±) is the cutoff that determines how strong the evidence must be before rejecting the null hypothesis.


Question 4: What are Type I and Type II Errors? Give Examples of Each.



## üîπ Type I Error (False Positive)

A **Type I error** occurs when we **reject the null hypothesis (H‚ÇÄ) even though it is actually true**.

### ‚úî Example:
A medical test says a person **has a disease**, but the person is actually **healthy**.

- Reality: H‚ÇÄ is true (no disease)
- Decision: Reject H‚ÇÄ (test says disease)
- Error: False alarm

This error rate is controlled by the **significance level (Œ±)**.

---

## üîπ Type II Error (False Negative)

A **Type II error** happens when we **fail to reject the null hypothesis when the alternative hypothesis (H‚ÇÅ) is actually true**.

### ‚úî Example:
A medical test says a person **does not have a disease**, but the person **actually has it**.

- Reality: H‚ÇÅ is true (disease present)
- Decision: Fail to reject H‚ÇÄ (test says no disease)
- Error: Missed detection

This error rate is denoted by **Œ≤**, and (1 ‚àí Œ≤) is the **power of the test**.

---

## üîπ Summary Table

| Error Type | Occurs When | Meaning | Example |
|-----------|-------------|----------|----------|
| **Type I (Œ±)** | Reject H‚ÇÄ when it is true | False positive | Healthy person diagnosed as sick |
| **Type II (Œ≤)** | Fail to reject H‚ÇÄ when it is false | False negative | Sick person diagnosed as healthy |

---

## ‚úîÔ∏è In one line:
- **Type I Error = False positive**  
- **Type II Error = False negative**


Question 5: What is the Difference Between a Z-test and a T-test? When to Use Each?

## üîπ Z-test

A **Z-test** is used when:
- The **sample size is large** (n > 30)
- The **population standard deviation (œÉ)** is known
- The data is approximately **normally distributed**

### ‚úî Use Z-test when:
- Testing mean with large samples  
- Population variance is known  
- Comparing sample mean to population mean (large n)

---

## üîπ T-test

A **T-test** is used when:
- The **sample size is small** (n ‚â§ 30)
- The **population standard deviation is unknown**  
- Sample variance (s) is used instead  
- Data is normally distributed

### ‚úî Use T-test when:
- Testing mean with **small samples**  
- Population SD is unknown  
- Comparing two sample means (independent or paired)

---

## üîπ Key Differences

| Feature | Z-test | T-test |
|--------|--------|--------|
| Sample Size | Large (n > 30) | Small (n ‚â§ 30) |
| Population SD | Known | Unknown |
| Test Statistic | Uses Z distribution | Uses t distribution |
| Spread | Less spread | More spread (fat tails) |
| When to use | Known variance, large samples | Unknown variance, small samples |

---

## ‚úîÔ∏è In one line:
Use a **Z-test** for large samples with known population SD, and a **T-test** for small samples with unknown population SD.


Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. (Include your Python code and output in the code box below.) Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10         # number of trials
p = 0.5        # probability of success
size = 1000    # number of random samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Print first few values
print("Generated Data (first 20 values):")
print(data[:20])

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9] (Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy.stats import norm

# Given sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Hypothesized population mean
mu_0 = 50

# Assume population standard deviation (œÉ)
sigma = 1    # You may adjust this if given

# Compute sample statistics
sample_mean = np.mean(data)
n = len(data)

# Z statistic
Z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(Z)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", Z)
print("p-value:", p_value)
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. (Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random normal data
np.random.seed(0)  # ensures reproducibility
mu = 50
sigma = 10
n = 1000

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate confidence interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
margin_of_error = 1.96 * (sample_std / np.sqrt(n))

lower_ci = sample_mean - margin_of_error
upper_ci = sample_mean + margin_of_error

# Print the results
print("Sample Mean:", sample_mean)
print("95% Confidence Interval: (", lower_ci, ",", upper_ci, ")")

# Step 3: Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Simulated Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. (Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std
    z_scores = (data - mean) / std
    return z_scores, mean, std

# Example dataset
np.random.seed(0)
data = np.random.normal(100, 15, 500)  # mean=100, std=15

# Calculate Z-scores
z_scores, mean, std = calculate_z_scores(data)

print("Original Mean:", mean)
print("Original Standard Deviation:", std)
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.figure(figsize=(8,5))
plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
