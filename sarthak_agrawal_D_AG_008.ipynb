## Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.


**Answer:**

Simple Linear Regression (SLR) is a supervised learning algorithm used to model the linear relationship between one independent variable (X) and one dependent variable (Y).

**Purpose:**
- To predict the dependent variable
- To understand the relationship between variables
- To analyze trends in data


## Question 2: What are the key assumptions of Simple Linear Regression?


**Answer:**

1. Linearity – Linear relationship between X and Y  
2. Independence – Observations are independent  
3. Homoscedasticity – Constant variance of errors  
4. Normality – Errors are normally distributed  
5. Single predictor – Only one independent variable


## Question 3: Write the mathematical equation for a simple linear regression model and explain each term.


**Answer:**

\[
y = \beta_0 + \beta_1 x + \epsilon
\]

Where:  
- \(y\): Dependent variable  
- \(x\): Independent variable  
- \(\beta_0\): Intercept  
- \(\beta_1\): Slope  
- \(\epsilon\): Error term


## Question 4: Provide a real-world example where simple linear regression can be applied.


**Answer:**

Predicting house prices based on house size.

- X: House size (square feet)
- Y: House price


## Question 5: What is the method of least squares in linear regression?


**Answer:**

The method of least squares finds the best-fit regression line by minimizing the sum of squared differences between actual and predicted values.

\[
\sum (y_{actual} - y_{predicted})^2
\]


## Question 6: What is Logistic Regression? How does it differ from Linear Regression?


**Answer:**

Logistic Regression is used for binary classification problems where the output is 0 or 1.

| Feature | Linear Regression | Logistic Regression |
|------|------------------|---------------------|
| Output | Continuous | Binary |
| Use Case | Regression | Classification |
| Function | Linear | Sigmoid |


## Question 7: Name and briefly describe three common evaluation metrics for regression models.


**Answer:**

1. Mean Absolute Error (MAE) – Average absolute error  
2. Mean Squared Error (MSE) – Average squared error  
3. Root Mean Squared Error (RMSE) – Square root of MSE


## Question 8: What is the purpose of the R-squared metric in regression analysis?


**Answer:**

R-squared (\(R^2\)) measures how much variance in the dependent variable is explained by the model.

- Value ranges from 0 to 1
- Higher value indicates better model fit


## Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.


```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

# Model training
model = LinearRegression()
model.fit(X, y)

# Output
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

## Question 10: How do you interpret the coefficients in a simple linear regression model?


**Answer:**

- Intercept (\(\beta_0\)) represents the value of Y when X is zero  
- Slope (\(\beta_1\)) represents the change in Y for a one-unit change in X
