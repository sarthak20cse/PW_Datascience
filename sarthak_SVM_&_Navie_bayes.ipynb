Q1. What is a Support Vector Machine (SVM)?

A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression. It finds an optimal hyperplane that maximizes the margin between different classes in the feature space.

Q2.What is the difference between Hard Margin and Soft Margin SVM
| Hard Margin                             | Soft Margin              |
| --------------------------------------- | ------------------------ |
| No misclassification allowed            | Allows misclassification |
| Works only for perfectly separable data | Works for noisy data     |
| Sensitive to outliers                   | Robust to outliers       |

Q4. What is the role of Lagrange Multipliers in SVM

Lagrange multipliers convert the constrained optimization problem into a dual problem, enabling efficient computation and kernel usage.

Q5. What are Support Vectors in SVM?

Support Vectors are the data points closest to the decision boundary.
They define the position and orientation of the hyperplane‚Äîremoving them changes the model.

Q6. What is a Support Vector Classifier (SVC)

A Support Vector Classifier (SVC) is the classification implementation of SVM that uses kernel functions to classify data into different classes, including non-linearly separable data.

Q7. What is a Support Vector Regressor (SVR)

Support Vector Regressor (SVR) is the regression version of SVM that fits a function within a predefined error margin while maintaining maximum margin.

Q8. What is the Kernel Trick in SVM

The Kernel Trick allows SVM to handle non-linearly separable data by implicitly mapping input data into a higher-dimensional space without explicitly computing the transformation.

Q9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel

The Linear Kernel is used for linearly separable data. The Polynomial Kernel is suitable for moderately non-linear data. The RBF Kernel is used for highly non-linear data and is the most widely used kernel due to its flexibility.

Q10. What is the effect of the C parameter in SVM

The C parameter controls the trade-off between maximizing the margin and minimizing classification error. A high value of C leads to less tolerance for misclassification, while a low value allows more misclassification.

Q11. What is the role of the Gamma parameter in RBF Kernel SVM

Gamma controls the influence of individual training points. A high gamma value leads to overfitting, while a low gamma value leads to underfitting.

Q12. What is the Na√Øve Bayes classifier, and why is it called "Na√Øve"

Na√Øve Bayes is a probabilistic classifier based on Bayes‚Äô Theorem. It is called "Na√Øve" because it assumes that all features are independent of each other.

Q14. Explain the differences between Gaussian Na√Øve Bayes, Multinomial Na√Øve Bayes, and Bernoulli Na√Øve Bayes

Gaussian Na√Øve Bayes is used for continuous data, Multinomial Na√Øve Bayes is used for count-based data such as text, and Bernoulli Na√Øve Bayes is used for binary features.

Q15. When should you use Gaussian Na√Øve Bayes over other variants

Gaussian Na√Øve Bayes should be used when the features are continuous and approximately follow a normal distribution.

Q16. What are the key assumptions made by Na√Øve Bayes

Na√Øve Bayes assumes feature independence, equal importance of features, and that the data follows a known probability distribution.

Q17. What are the advantages and disadvantages of Na√Øve Bayes

Na√Øve Bayes is fast, simple, and efficient for large datasets but performs poorly when features are highly correlated.

Q18. Why is Na√Øve Bayes a good choice for text classification

Na√Øve Bayes performs well in text classification because it handles high-dimensional sparse data efficiently and requires less training data.

Q19. Compare SVM and Na√Øve Bayes for classification tasks

SVM generally provides higher accuracy and handles complex boundaries, while Na√Øve Bayes is faster, simpler, and works well with large text datasets.

Q20. How does Laplace Smoothing help in Na√Øve Bayes?

Laplace Smoothing prevents zero probability issues by adding a small constant to all feature counts, ensuring stable probability estimates.

Q21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

Q22. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = load_wine(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

linear = SVC(kernel='linear')
rbf = SVC(kernel='rbf')

linear.fit(X_train, y_train)
rbf.fit(X_train, y_train)

print("Linear Accuracy:", accuracy_score(y_test, linear.predict(X_test)))
print("RBF Accuracy:", accuracy_score(y_test, rbf.predict(X_test)))
```

Q23. Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean Squared Error (MSE)

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

X, y = fetch_california_housing(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)

y_pred = svr.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
```

Q24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

X, y = make_blobs(n_samples=100, centers=2)
model = SVC(kernel='poly', degree=3)
model.fit(X, y)

xx, yy = np.meshgrid(np.linspace(-5,5,200), np.linspace(-5,5,200))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

Q26. Write a Python program to train a Multinomial Na√Øve Bayes classifier for text classification using the 20 Newsgroups dataset

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

data = fetch_20newsgroups()
X = CountVectorizer().fit_transform(data.data)
y = data.target

model = MultinomialNB()
model.fit(X, y)

print("Accuracy:", accuracy_score(y, model.predict(X)))
```

Q27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

X, y = make_blobs(n_samples=100, centers=2)

for C in [0.1, 1, 10]:
    model = SVC(C=C, kernel='linear')
    model.fit(X, y)
    plt.scatter(X[:,0], X[:,1], c=y)
    plt.title(f"C = {C}")
    plt.show()
```

Q29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X, y = load_wine(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

model1 = SVC()
model1.fit(X_train, y_train)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train_s, X_test_s, y_train, y_test = train_test_split(X_scaled, y)

model2 = SVC()
model2.fit(X_train_s, y_train)

print("Unscaled Accuracy:", model1.score(X_test, y_test))
print("Scaled Accuracy:", model2.score(X_test_s, y_test))
```

Q31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

params = {'C':[1,10], 'gamma':[0.1,1], 'kernel':['rbf']}
grid = GridSearchCV(SVC(), params)
grid.fit(X, y)

print("Best Parameters:", grid.best_params_)
```

Q38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy

Q40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
rfe = RFE(SVC(kernel='linear'), n_features_to_select=2)
X_new = rfe.fit_transform(X, y)

model = SVC()
model.fit(X_new, y)

print("Accuracy:", model.score(X_new, y))
```

Q45. Write a Python program to train a Na√Øve Bayes classifier and evaluate its performance using the ROC-AUC score


```python
from sklearn.metrics import roc_auc_score
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_breast_cancer

X, y = load_breast_cancer(return_X_y=True)
model = GaussianNB().fit(X, y)

print("ROC-AUC:", roc_auc_score(y, model.predict_proba(X)[:,1]))
```

Q46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve

```python
from sklearn.metrics import precision_recall_curve
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

X, y = load_breast_cancer(return_X_y=True)
model = SVC(probability=True).fit(X, y)

precision, recall, _ = precision_recall_curve(y, model.predict_proba(X)[:,1])
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```
