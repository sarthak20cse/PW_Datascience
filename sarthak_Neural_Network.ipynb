Question 1
What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

Answer:

Deep Learning is a subset of machine learning that uses artificial neural networks with multiple layers to learn patterns from large amounts of data. These networks are inspired by the structure and functioning of the human brain. Deep learning models automatically learn hierarchical representations of data, which makes them very effective for tasks such as image recognition, speech recognition, natural language processing, and recommendation systems.

The evolution of deep learning started from early neural network research in the 1950s with the perceptron model. However, due to computational limitations and lack of large datasets, progress was slow. In the 1980s and 1990s, improvements such as backpropagation helped neural networks train more effectively. The real breakthrough occurred after 2010 due to three major factors: availability of large datasets, powerful GPUs, and improved algorithms. This enabled the development of deep neural networks capable of solving complex problems.

Deep learning differs from traditional machine learning in several ways. Traditional machine learning methods usually require manual feature engineering where domain experts design features from raw data. In contrast, deep learning automatically extracts features through multiple neural network layers. Additionally, traditional machine learning algorithms like decision trees or support vector machines generally work well with smaller datasets, while deep learning performs best with very large datasets. Deep learning models are also computationally more intensive but can achieve higher accuracy for complex tasks.

Question 2
Explain the basic architecture and functioning of a Perceptron. What are its limitations?

Answer:

A perceptron is the simplest type of artificial neural network and serves as the basic building block of modern neural networks. It is a single-layer binary classification model that takes multiple input values and produces a single output.

The architecture of a perceptron consists of:

Input layer: Receives input features

Weights: Each input is associated with a weight

Bias: A constant value added to the weighted sum

Activation function: Determines the final output

The perceptron works by multiplying each input with its corresponding weight, summing them together, and adding a bias value. This result is then passed through an activation function such as a step function. If the result exceeds a certain threshold, the perceptron outputs 1; otherwise, it outputs 0.

Mathematically:

y = activation( w₁x₁ + w₂x₂ + ... + wₙxₙ + b )

During training, the perceptron updates its weights using the perceptron learning rule based on prediction error.

Limitations of Perceptron:

It can only classify linearly separable data.

It cannot solve problems like XOR that are not linearly separable.

It has only one layer and cannot learn complex patterns.

It uses a simple activation function which limits learning capability.

These limitations led to the development of multilayer neural networks.

Question 3
Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

Answer:

Activation functions introduce non-linearity into neural networks, enabling them to learn complex patterns from data. Without activation functions, neural networks would behave like simple linear models regardless of the number of layers.

The activation function determines whether a neuron should be activated based on the weighted sum of inputs.

Comparison of common activation functions:

1. Sigmoid Function

Formula:

σ(x) = 1 / (1 + e⁻ˣ)

Characteristics:

Output range: 0 to 1

Smooth and differentiable

Often used for binary classification

Limitations:

Suffers from vanishing gradient problem

Slow convergence

2. Tanh Function

Formula:

tanh(x) = (eˣ − e⁻ˣ) / (eˣ + e⁻ˣ)

Characteristics:

Output range: -1 to 1

Zero centered

Better gradient flow than sigmoid

Limitations:

Still suffers from vanishing gradient

3. ReLU (Rectified Linear Unit)

Formula:

f(x) = max(0, x)

Characteristics:

Output range: 0 to ∞

Very fast computation

Most widely used in deep learning

Limitations:

Dying ReLU problem (neurons may stop learning)

Question 4
What is the difference between Loss function and Cost function in neural networks? Provide examples.

Answer:

Loss function measures the error for a single training example, while cost function measures the average loss across the entire dataset.

Loss Function:
It calculates how far the predicted value is from the actual value for one data sample.

Example:

Mean Squared Error (MSE)

Binary Cross Entropy

Example formula:

Loss = (y_true − y_pred)²

Cost Function:
It is the average of loss values for all samples in the dataset.

Cost = (1/n) Σ Loss

Examples:

Mean Squared Error (Regression)

MSE = (1/n) Σ (y_true − y_pred)²

Binary Cross Entropy (Binary Classification)

Loss = −[ y log(p) + (1 − y) log(1 − p) ]

Categorical Cross Entropy (Multiclass Classification)

Cost functions guide the training process by minimizing prediction error during optimization.

Question 5
What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

Answer:

Optimizers are algorithms used to update the weights of neural networks in order to minimize the loss function during training.

They determine how the model learns from errors and improves its predictions.

1. Gradient Descent

Basic optimization algorithm that updates weights in the opposite direction of the gradient.

Update rule:

w = w − α * gradient

Advantages:

Simple and widely used

Disadvantages:

Slow convergence

Sensitive to learning rate

2. RMSprop

Improves gradient descent by adapting the learning rate for each parameter.

Advantages:

Faster convergence

Works well for non-stationary problems

Disadvantages:

Slightly more complex

3. Adam (Adaptive Moment Estimation)

Combines RMSprop and Momentum techniques.

Advantages:

Fast convergence

Works well for most deep learning tasks

Adaptive learning rates

Disadvantages:

Sometimes overfits

Adam is currently one of the most widely used optimizers.

Question 6
Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

```python
import numpy as np

# Input dataset (AND gate)
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Output
y = np.array([0,0,0,1])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1

# Activation function
def step(x):
    return 1 if x >= 0 else 0

# Training
for epoch in range(10):
    for i in range(len(X)):
        linear = np.dot(X[i], weights) + bias
        y_pred = step(linear)

        error = y[i] - y_pred

        weights += learning_rate * error * X[i]
        bias += learning_rate * error

print("Weights:", weights)
print("Bias:", bias)

# Testing
for x in X:
    result = step(np.dot(x, weights) + bias)
    print(x, "->", result)
```

Question 7
Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0,x)
tanh = np.tanh(x)

plt.figure(figsize=(8,6))

plt.plot(x,sigmoid,label="Sigmoid")
plt.plot(x,relu,label="ReLU")
plt.plot(x,tanh,label="Tanh")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()

plt.show()
```

Question 8
Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load dataset
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()

# Normalize
X_train=X_train/255.0
X_test=X_test/255.0

# Build model
model=keras.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128,activation='relu'),
    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history=model.fit(X_train,y_train,epochs=5)

print("Training Accuracy:", history.history['accuracy'][-1])
```

Question 9
Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

(X_train,y_train),(X_test,y_test)=keras.datasets.fashion_mnist.load_data()

X_train=X_train/255.0
X_test=X_test/255.0

model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history=model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))

# Plot accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.legend(["Train","Validation"])
plt.show()

# Plot loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Loss")
plt.legend(["Train","Validation"])
plt.show()
```

You are working on a project for a bank that wants to automatically detect fraudulent transactions.

Answer:

To solve the fraud detection problem, I would design a multilayer neural network that takes transaction features such as amount, merchant ID, transaction time, and customer location as inputs. The model would consist of multiple dense layers with nonlinear activation functions to learn complex patterns in transaction behavior.

For activation functions, ReLU would be used in hidden layers because it helps neural networks learn faster and avoids vanishing gradients. For the output layer, a sigmoid activation function would be used because the task is a binary classification problem (fraud or legitimate).

The loss function used would be binary cross entropy because it works well for binary classification problems.

Since the dataset is imbalanced, techniques such as class weighting, oversampling using SMOTE, or undersampling would be used to balance the training data. Evaluation metrics such as precision, recall, F1-score, and ROC-AUC would be used instead of accuracy.

The Adam optimizer would be used because it adapts learning rates automatically and converges faster.

To prevent overfitting, techniques such as dropout layers, early stopping, and L2 regularization would be applied.

```python
import tensorflow as tf
from tensorflow.keras import layers

model=tf.keras.Sequential([
    layers.Dense(64,activation='relu',input_shape=(10,)),
    layers.Dropout(0.3),
    layers.Dense(32,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print(model.summary())
```
