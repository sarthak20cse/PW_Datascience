Question 1:  What is the difference between descriptive statistics and inferential statistics? Explain with examples.

1. Descriptive Statistics

Descriptive statistics are used to summarize, organize, and describe the main features of a dataset.
It deals with actual data and does not attempt to make predictions or generalizations beyond the data collected.
Common tools used include mean, median, mode, range, standard deviation, percentage, graphs, and tables.

Example:
If we collect marks of 50 students in a class and calculate their average marks, highest and lowest marks, and represent them in a bar graph, this is descriptive statistics. It only describes what the data shows.

2. Inferential Statistics

Inferential statistics are used to draw conclusions, make predictions, or test hypotheses about a larger population based on a smaller sample of data.
It helps to infer results beyond the data at hand using methods like hypothesis testing, confidence intervals, correlation, regression, and ANOVA.

Example:
If we take the marks of 50 students from one class (sample) and use them to estimate the average marks of all students in the school (population), we are using inferential statistics. It helps to make generalizations from sample data.

Difference between Descriptive and Inferential Statistics

Meaning:

Descriptive statistics describe and summarize the data collected.

Inferential statistics make predictions or inferences about a population from a sample.

Purpose:

Descriptive statistics aim to present data in a clear and understandable form.

Inferential statistics aim to generalize results and test hypotheses.

Data Used:

Descriptive statistics use complete or observed data.

Inferential statistics use sample data to draw conclusions about a larger population.

Techniques Used:

Descriptive uses measures like mean, median, mode, charts, and graphs.

Inferential uses statistical tests like t-tests, regression, and confidence intervals.

Example:

Descriptive: Finding the average marks of students in a class.

Inferential: Predicting the average marks of all students in a school from a class sample.

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

In statistics, sampling is the process of selecting a small group (sample) from a larger group (population) to collect data and draw conclusions about the entire population.
It is used when studying the whole population is difficult, time-consuming, or expensive.
A sample should represent the population accurately so that the results can be generalized.

Example:

If a company wants to know the satisfaction level of its 10,000 customers, it can survey 500 customers instead of all 10,000. The 500 customers form the sample, and the 10,000 customers form the population.

Difference between Random and Stratified Sampling

Definition:

Random sampling selects individuals purely by chance.

Stratified sampling divides the population into groups and samples are taken from each group.

Representation:

Random sampling may not represent all subgroups well.

Stratified sampling ensures representation from all subgroups.

Complexity:

Random sampling is simple to conduct.

Stratified sampling is more complex and requires prior knowledge of population characteristics.

Use Case:

Random sampling is used when the population is homogeneous (similar).

Stratified sampling is used when the population is heterogeneous (diverse).

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.



In statistics, **Mean**, **Median**, and **Mode** are the three main **measures of central tendency**.  
They are used to find the **center or average value** around which the data values are distributed.  
These measures help in summarizing large data into a single representative value.

---

## **1. Mean**
- The **mean** is the **arithmetic average** of a set of numbers.  
- It is calculated by adding all the observations and dividing the sum by the total number of observations.  

**Formula:**
\[
\text{Mean} = \frac{\text{Sum of all observations}}{\text{Total number of observations}}
\]

**Example:**
If the marks of 5 students are 60, 70, 80, 90, and 100, then  
\[
\text{Mean} = \frac{60 + 70 + 80 + 90 + 100}{5} = \frac{400}{5} = 80
\]

---

## **2. Median**
- The **median** is the **middle value** of a dataset when the data is arranged in **ascending or descending order**.  
- If the number of observations is **odd**, the middle value is the median.  
- If the number of observations is **even**, the median is the **average of the two middle values**.  

**Example:**
- For the data 20, 25, 30, 35, 40 → **Median = 30**  
- For the data 10, 20, 30, 40 → **Median = (20 + 30) / 2 = 25**

---

## **3. Mode**
- The **mode** is the value that occurs **most frequently** in a dataset.  
- A dataset can have **no mode**, **one mode (unimodal)**, or **more than one mode (bimodal or multimodal)**.

**Example:**
For the data 5, 8, 8, 10, 12, 8, 15 → **Mode = 8**

---

## **Importance of Measures of Central Tendency**

1. **Data Summarization:**  
   They condense a large set of data into a single representative value, making it easier to understand.

2. **Comparison:**  
   They help compare different datasets (e.g., comparing average income or average marks).

3. **Decision Making:**  
   In business, economics, and research, these measures guide decisions based on average performance or outcomes.

4. **Foundation for Further Analysis:**  
   Many statistical techniques, such as variance and standard deviation, are based on the mean.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?


In statistics, **skewness** and **kurtosis** are measures that describe the **shape** and **distribution** of data.  
They help us understand how the data deviates from a normal (symmetrical) distribution.

---

## **1. Skewness**

**Definition:**  
Skewness measures the **asymmetry** of a distribution.  
In a perfectly symmetrical distribution (like a normal curve), the **mean, median, and mode** are equal, and **skewness = 0**.

### **Types of Skewness:**

- **a) Positive Skew (Right Skewed):**  
  - The **tail of the distribution extends to the right**.  
  - Most of the data values are concentrated on the **left side**, and a few large values pull the mean to the right.  
  - **Mean > Median > Mode**

  **Example:** Income distribution — most people earn low or average incomes, while a few earn very high incomes.

- **b) Negative Skew (Left Skewed):**  
  - The **tail of the distribution extends to the left**.  
  - Most of the data values are concentrated on the **right side**, and a few small values pull the mean to the left.  
  - **Mean < Median < Mode**

  **Example:** Age at retirement — most people retire around the same age, but a few retire much earlier.

---

## **2. Kurtosis**

**Definition:**  
Kurtosis measures the **peakedness** or **flatness** of a data distribution compared to the normal distribution.  
It indicates how heavy or light the tails of the distribution are.

### **Types of Kurtosis:**

- **a) Mesokurtic:**  
  - Normal distribution (standard bell curve).  
  - Moderate peak, kurtosis = 3.

- **b) Leptokurtic:**  
  - Distribution has a **sharper peak** and **heavier tails**.  
  - More values are concentrated around the mean, but extreme values (outliers) are more likely.  
  - Kurtosis > 3.

- **c) Platykurtic:**  
  - Distribution has a **flatter peak** and **lighter tails**.  
  - Data is more spread out.  
  - Kurtosis < 3.

---

## **3. Interpretation of Positive Skew**

A **positive skew** (right-skewed distribution) means:
- The **tail extends to the right** side.  
- The **mean is greater than the median**.  
- The data contains a few **very large values (outliers)** that pull the average upward.  
- Most of the observations are **below the mean**.

**Example:**  
If most people earn ₹20,000–₹50,000 per month but a few earn ₹5,00,000 or more, the income distribution is **positively skewed**.

---

## **Conclusion**

- **Skewness** describes the **direction and extent of asymmetry** in data.  
- **Kurtosis** describes the **height and heaviness** of the distribution’s tails.  
- A **positive skew** indicates that the data has a longer right tail, with most values being smaller and a few extreme high values influencing the mean.


Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] (Include your Python code and output in the code box below.)

```python
# **Program to compute Mean, Median, and Mode**

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Importing statistics module
import statistics as stats

# Calculating Mean, Median, and Mode
mean_value = stats.mean(numbers)
median_value = stats.median(numbers)
mode_value = stats.mode(numbers)

# Displaying results
print("Given Numbers:", numbers)
print("Mean  =", mean_value)
print("Median =", median_value)
print("Mode  =", mode_value)
```

Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] (Include your Python code and output in the code box below.)

```python
# **Program to compute Covariance and Correlation Coefficient**

# Given data lists
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Importing necessary libraries
import numpy as np

# Calculating means
mean_x = np.mean(list_x)
mean_y = np.mean(list_y)

# Computing Covariance
covariance = np.cov(list_x, list_y, bias=True)[0][1]

# Computing Correlation Coefficient
correlation = np.corrcoef(list_x, list_y)[0][1]

# Displaying results
print("List X:", list_x)
print("List Y:", list_y)
print("Mean of X =", mean_x)
print("Mean of Y =", mean_y)
print("Covariance =", covariance)
print("Correlation Coefficient =", correlation)
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] (Include your Python code and output in the code box below.)

```python
# **Program to draw a Boxplot and identify outliers**

import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Creating the boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Given Data")
plt.xlabel("Values")
plt.show()

# Calculating IQR (Interquartile Range) to identify outliers
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Defining outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Finding outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

### Relationship between Advertising Spend and Daily Sales

As a data analyst, to understand the relationship between **advertising spend** and **daily sales**, we can use:
- **Covariance** to see if both variables move together (positive or negative direction).
- **Correlation** to measure the *strength* and *direction* of this relationship on a standardized scale (-1 to +1).


```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute covariance
covariance = np.cov(advertising_spend, daily_sales, ddof=1)[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

### Analyzing Customer Satisfaction Survey Data

To understand the distribution of customer satisfaction scores (1–10 scale), we can use:

**Summary Statistics:**
- **Mean:** Measures the average satisfaction level.
- **Median:** Represents the middle score, less affected by outliers.
- **Mode:** Shows the most common satisfaction score.
- **Standard Deviation:** Indicates how spread out the scores are.

**Visualizations:**
- **Histogram:** Displays how frequently each score occurs — helps identify skewness and spread.
- **Boxplot (optional):** Shows median, quartiles, and potential outliers.

These tools help us understand if most customers are satisfied (high scores) or if opinions vary widely.


```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, median, mode, stdev

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
print("Mean:", mean(survey_scores))
print("Median:", median(survey_scores))
print("Mode:", mode(survey_scores))
print("Standard Deviation:", round(stdev(survey_scores), 2))

# Plot histogram
plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Score Distribution")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
